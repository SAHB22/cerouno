```python
# Sustituye el nombre dependiendo de la posición de su columna
Guarda_Paises.rename(columns = {list(Guarda_Paises)[0]:'Ubicación'}, inplace=True)
Guarda_Paises.rename(columns = {list(Guarda_Paises)[1]:'Pais'}, inplace=True)

Guarda_Paises.rename(columns={'INDICATOR': 'INDICADOR', 'Indicator': 'Indicador', "MEASURE" : "MEDIBLE","Measure":"Medible"}, inplace=True)
Guarda_Paises.rename(columns={'INEQUALITY': 'INEQUIDAD',"Inequality":"Inequidad", 'Unit Code': 'Código de Unidad', "Unit" : "Unidad"}, inplace=True)
Guarda_Paises.rename(columns={'PowerCode Code': 'Poder de Código Código', 'PowerCode': 'Poder de Código'}, inplace=True)
Guarda_Paises.rename(columns={'Reference Period Code': 'El periodo de código Referente', 'Reference Period': 'Periodo de referencia', "Value" : "Valor", "Flag Codes":"Códigos de Bandera","Flags":"Banderas"}, inplace=True)
Guarda_Paises.head()
```
